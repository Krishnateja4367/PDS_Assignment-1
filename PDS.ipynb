```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cleandata=pd.read_csv('1stquescleandata.csv')
```

```python
cleandata
```

```python
sns.scatterplot(x='Height (Inches)', y='Age', data=cleandata)
plt.show()
```

```python
sns.lineplot(x='Age', y='Grip strength', data=cleandata)
```

```python
data = cleandata[['Height (Inches)', 'Age', 'Frailty']]
correlation_matrix = data.corr()
sns.heatmap(data=correlation_matrix, annot=True)
plt.show()
```
