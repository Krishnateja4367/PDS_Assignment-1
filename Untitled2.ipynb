```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw=pd.read_csv('StudentsPerformance.csv')
```

```python
raw
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='race/ethnicity', y='reading score', data=raw)
plt.title('Reading Score Distribution by Race/Ethnicity')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.pairplot(raw[['math score', 'reading score', 'writing score']])
plt.title('Pairwise Relationships between Scores')
plt.show()
```

```python
plt.figure(figsize=(10,6))
correlation_matrix = raw[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(data=correlation_matrix, annot=True)
plt.title('Correlation between Scores')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(raw['writing score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Writing Scores')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(raw['math score'], raw['reading score'], alpha=0.5)
plt.title('Math vs Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```
